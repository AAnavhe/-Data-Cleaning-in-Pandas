## Dataset Cleaning

```python
import numpy as np
import pandas as pd 
```

```python
df= pd.read_csv(r"C:\Users\EliteBook 840 G3\OneDrive\messy_data.csv")
df
#'df' is data frame.
```

```python
df.shape
```

```python
df[df.TrnsnID.duplicated()]
# Remove duplicate records i.e 'TrnsnID'
```

```python
df.drop_duplicates()
#Delete duplicate value(S) in 'TrnsnID'
```

```python
df.shape
```

```python
df.isna().sum()
#check for missing values
```

```python
df['ProdtName'].bfill(inplace = True) 
df['PaymentType'].bfill(inplace = True)
df.isna().sum()
```

```python
df['Data'] = pd.to_datetime(df['Data'], format ='mixed')
df.sample(6)
#Standardize date format
```

```python
df['PaymentType'].replace({'debit card':'Debit Card','cash': 'Cash','Debit':'Debit Card','Credit Card':'Credit Card','Credit':'Credit Card', 'paypal':'Paypal'}, inplace = True)
df.sample(10)
#Standardize payment type names
```

```python
df.PaymentType.value_counts()
```

```python
df['CusterAge']= np.abs(df['CusterAge'])
df.sample(6)
#Check for negative values in 'CusterAge' and handle them accordingly
```

```python
df['Qutity']= np.abs(df['Qutity'])
df.sample(6)
#Check for negative values in 'Qutity' and handle them accordingly
```

```python
df['ProdtName'].replace({'Boo':'Book','Bok':'Book', 'Bookk':'Book', }, regex = True, inplace = True)
df['ProdtName'].replace('kk', 'k', regex=True, inplace = True)
df
#Correct typos in product names
```

```python
df['TtlSales'] = df['TtlSales'].astype('float64')
df.head(10)
#Ensure correct data types
```

```python
df.columns
```

```python
df.rename(columns={'TrnsnID':'TransactionID', 'Data':'Date', 'CusmrID':'CustomerID', 'ProdtID':'ProductID', 'ProdtName':'ProductName', 'Qutity':'Quantity', 'Prize':'Price', 'TtlSales':'TotalSales', 'CusterAge': 'CustomerAge'},)
#Correct misspelled column names
```
